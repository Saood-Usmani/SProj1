```python
from sklearn.model_selection import train_test_split
X = data.drop('price_above_median', axis=1)
y = data['price_above_median']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=1)
```

```python
#will need this class to set up advanced pipeline
from sklearn.base import BaseEstimator
from sklearn.neighbors import KNeighborsClassifier

class MultiModelClassifier(BaseEstimator):
    """
    A custom Estimator class that can be constructed with different model types.
    For details on implementing custom Estimators,
    see: https://scikit-learn.org/stable/developers/develop.html
    """

    def __init__(self, model=KNeighborsClassifier()):
        """
        A custom estimator parameterized by the model.
        Pass the result of an estimator constructor for `model`. By default,
        it uses the KNeighborsClassifier().
        """
        self.model = model

    def fit(self, X, y=None, **kwargs):
        self.model.fit(X, y)
        return self

    def predict(self, X, y=None):
        return self.model.predict(X)

    def predict_proba(self, X):
        return self.model.predict_proba(X)

    def score(self, X, y):
        return self.model.score(X, y)
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline

p2 = Pipeline([
    ('scale', RobustScaler()),
    ('mmc', MultiModelClassifier()),
])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC


import numpy as np
param_grid = [
    {
        "mmc__model": [KNeighborsClassifier()],
        "mmc__model__n_neighbors": np.arange(1, 100, step = 5)
    },
    {
        "mmc__model": [DecisionTreeClassifier()],
        "mmc__model__max_depth": np.arange(start=2, stop=20),
        "mmc__model__min_samples_leaf": np.arange(start=1, stop=5),
    },
    {
        "mmc__model": [AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1))], 
        "mmc__model__learning_rate": [0.1, 0.5, 1.0],
        "mmc__model__n_estimators": [50, 100, 200]
        # "mmc__model__n_estimators": np.arange(start=10, stop=300, step=1),
    },
    {
        "mmc__model": [RandomForestClassifier()],
        "mmc__model__n_estimators": np.arange(start=10, stop=200),
        # "mmc__model__max_depth": np.arange(start=2, stop=20),
        # "mmc__model__min_samples_leaf": np.arange(start=1, stop=5),
        # "mmc__model__class_weight": [{0: 0.1, 1: 0.9}, {0: 0.2, 1: 0.8}, {0: 0.3, 1: 0.7}],
    },
    # {
    #     "mmc__model": [SVC()]
    # }
]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
gscv2 = GridSearchCV(p2, param_grid, cv=5, n_jobs=8, scoring = "precision")
gscv2.fit(X_train, y_train)
```

```python
#code generated with help of chatGPT
from collections import defaultdict
from sklearn.metrics import classification_report

# Dictionary to store models and their scores
model_scores = defaultdict(list)
results = gscv2.cv_results_

# Loop through results
for i, params in enumerate(results["params"]):
    model_name = str(params["mmc__model"].__class__.__name__)  # Extract model name
    mean_test_score = results["mean_test_score"][i]  # Extract corresponding score
    
    # Store in dictionary
    model_scores[model_name].append((params, mean_test_score))

# Sort each model type by score and select the best one
best_models = []
for model_name, models in model_scores.items():
    models.sort(key=lambda x: x[1], reverse=True)  # Sort by test score (descending)
    best_models.append((model_name, models[0]))  # Select the best model from each

# Sort the best models and select the top four
best_models.sort(key=lambda x: x[1][1], reverse=True)  # Sort by best model's score
top_four = best_models[:4]
best_pipeline = gscv2.best_estimator_
# Print results with classification reports
for model_name, (best_params, best_score) in top_four:
    # Initialize model with best parameters
    best_model = best_pipeline.set_params(**best_params)

    # Fit the model to training data (if not already fitted)
    best_model.fit(X_train, y_train)

    # Predict on test data
    y_test_pred = best_model.predict(X_test)
    # Print the best model details
    print(f"Model: {model_name}, Score: {best_score}, Params: {best_params}")

    # Print classification report
    print("\nClassification Report:\n")
    print(classification_report(y_test, y_test_pred))
    print("="*80)  # Separator for readability
```

```python
from sklearn.metrics import confusion_matrix
print(pd.DataFrame(confusion_matrix(y_test, gscv2.best_estimator_.predict(X_test)),
                columns=['Predicted_False', 'Predicted_True'],
                index=['Actual_False', 'Actual_True']))
```
